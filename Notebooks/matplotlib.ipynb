# Plotting with Matplotlib

Matplotlib is the defactor plotting library for Python. <br>
[Matplotlib](https://matplotlib.org)

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 100, 20)
z = np.linspace(-100, 10, 20)

[X, Z] = np.meshgrid(x, z)

g = 9.8

rho0 = 1000
drho_dx = 1/10e3
p_atm = 1e5

p = p_atm - g*(rho0 + drho_dx*X)*Z;



fig, ax = plt.subplots(1,1)
ax.contour(X, Z, p,10, cmap='PuBuGn')
```
