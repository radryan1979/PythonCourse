````python
fig, ax = plt.subplots()
ax.set_title('This Is My Graph')
ax.set_xlabel('The X Axes')
ax.set_ylabel('The Y Axes')
````

````python
x = np.linspace(0, 100, 20)
z = np.linspace(-100, 10, 20)

[X, Z] = np.meshgrid(x, z)

g = 9.8

rho0 = 1000
drho_dx = 1/10e3
p_atm = 1e5

p = p_atm - g*(rho0 + drho_dx*X)*Z;



fig, ax = plt.subplots(1,1)
ax.contour(X, Z, p,10, cmap='PuBuGn')
ax.set_title('Color Contours')
ax.set_xlabel('Whatever You Want')
ax.set_ylabel('Pressure')
````

An example of the stream plot from the Matplotlib website.

````python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery-nogrid')

# make a stream function:
X, Y = np.meshgrid(np.linspace(-3, 3, 256), np.linspace(-3, 3, 256))
Z = (1 - X/2 + X**5 + Y**3) * np.exp(-X**2 - Y**2)
# make U and V out of the streamfunction:
V = np.diff(Z[1:, :], axis=1)
U = -np.diff(Z[:, 1:], axis=0)

# plot:
fig, ax = plt.subplots()

ax.streamplot(X[1:, 1:], Y[1:, 1:], U, V)

plt.show()
````

Plotting a wave

````python
x = np.linspace(0, 20, 201)
y = np.sin(x)

plt.figure(figsize = (8, 6))
plt.plot(x, y, 'b')
plt.ylabel('Amplitude')
plt.xlabel('Location (x)')
plt.show()
````

````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np
fig = plt.figure()
ax = fig.gca(projection='3d')
x, y = np.meshgrid(np.arange(-2, 2, 0.2), np.arange(-2, 2, 0.2))
z = x*np.exp(-(x**2 + y**2))
surf = ax.plot_surface(x, y, z, cmap=cm.viridis)
plt.show()
````
