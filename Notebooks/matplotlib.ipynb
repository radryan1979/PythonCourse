```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()  # an empty figure with no Axes
fig, ax = plt.subplots()  # a figure with a single Axes
fig, axs = plt.subplots(2, 1)  # a figure with a 2x2 grid of Axes
```

```python
fig, ax = plt.subplots()
ax.set_title('This Is My Graph')
ax.set_xlabel('The X Axes')
ax.set_ylabel('The Y Axes')
```

```python
x = np.linspace(0, 100, 20)
z = np.linspace(-100, 10, 20)

[X, Z] = np.meshgrid(x, z)

g = 9.8

rho0 = 1000
drho_dx = 1/10e3
p_atm = 1e5

p = p_atm - g*(rho0 + drho_dx*X)*Z;



fig, ax = plt.subplots(1,1)
ax.contour(X, Z, p,10, cmap='PuBuGn')
ax.set_title('Color Contours')
ax.set_xlabel('Whatever You Want')
ax.set_ylabel('Pressure')
```
